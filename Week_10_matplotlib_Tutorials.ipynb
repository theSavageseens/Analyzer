**1- Write a Python program to draw a line with suitable label in the x axis, y axis and a title.**

```
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(1,30)
y = x**2

plt.plot(x,y)

plt.xlabel("x")
plt.ylabel("y")
plt.title("this is a plot")
plt.show()
```

**2- Write a Python program to plot two or more lines with legends, different widths and colors.**

```
x1 = np.arange(1,30)
y1 = x1**2


x2 = np.arange(1,30)
y2 = 1/2*x2

plt.xlabel("x")
plt.ylabel("y")
plt.title("this is a plot")

plt.plot(x1,y1, color='red', label = 'line1', linestyle='dotted')
plt.plot(x2,y2, color='blue', label = 'line2', linestyle='dashed')

plt.legend()
plt.show()
```

**3- Write a Python program to plot several lines with different format styles in one command using arrays.**

```
x1 = np.arange(1,30)
y1 = x1 * 2


x2 = np.arange(1,30)
y2 = x2 * 3

plt.xlabel("x")
plt.ylabel("y")
plt.title("this is a plot")

plt.plot(x1, y1, 'r--', x2, y2, 'b.', label=['line1', 'line2'])

plt.legend()
plt.show()
```

**4- Write a Python program to display a bar chart of the popularity of programming Languages.**

Programming languages: Java, Python, PHP, JavaScript, C#, C++

Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='blue')

plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")

plt.xticks(x_pos, x)

plt.show()
```

**5- Write a Python program to display a horizontal bar chart of the popularity of programming Languages.**

```
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = [i for i, _ in enumerate(x)]
plt.barh(x_pos, popularity, color='blue')

plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")

plt.yticks(x_pos, x)

plt.show()
```

**6- Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar.**

```
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color=['red', 'blue', 'green', 'orange', 'yellow', 'purple'])

plt.xlabel("Programming languages")
plt.ylabel("Popularity")
plt.title("popularity of programming Languages")

plt.xticks(x_pos, x)

plt.show()
```

**7- Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.**

```
x = np.random.randn(100) 
y = np.random.randn(100)
plt.scatter(x, y, s=50, facecolors='none', edgecolors='g')
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**8- Plot x in terms of y and y in terms of x in two different subplots of one figure.**

`x =[0, 1, 2, 3, 4, 5, 6]`

`y =[0, 1, 3, 6, 9, 12, 17]`

```
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 1, 3, 6, 9, 12, 17]

fig, axes = plt.subplots(1, 2) 

axes[0].plot(x, y, 'r--o') 
 
axes[1].plot(y, x, 'b--o') 

fig.tight_layout() 
```

**9- Plot x in terms of y and y in terms of x in 4 different colors and subplots of one figure. Use the data from the previous quesion.**

```
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) 

axes[0, 0].plot(x, y, 'r--o') 
axes[0, 0].set_title('red')

axes[0, 1].plot(y, x, 'b--o') 
axes[0, 1].set_title('blue')

axes[1, 0].plot(x, y, 'g--o') 
axes[1, 0].set_title('green')

axes[1, 1].plot(y, x, 'y--o') 
axes[1, 1].set_title('yellow')
  
fig.tight_layout() 
```
